# Define a function to be called later

```python
def read_file(file_path): #Declare a function called read_file

    with open(file_path,'r') as file: #Open file with permission to read 
        text = file.read() #Read the file
        print(text) #Print what is readed
        f = open("../combinedtranscript.txt", "a") #Open a file with permission to append
        f.write(text) #Write the text into the file
        f.write("\n \n \n \n") #Write couple blank lines between each transcript
        f.close() #Close the file
```

# Run the function & generate a "combinedtranscript.txt" file combining all individual transcripts

```python
f = open("../combinedtranscript.txt", "w") #These two lines create/clear a file called combinedtranscript.txt
f.close()
for file in os.listdir(path): 
    if file.endswith('.txt'): #If its a TXT file
        file_path = f"{path}/{file}" 
        read_file(file_path) #Execute the function defined earlier
```

```python
text_search = input("Enter keyword for search: ") #Ask for input as what keyword you wish to search
```

# Read lines in file and search for the keyword

```python
lines = file_read.readlines() #Read file in lines
new_list = [] #Create a blank new list
idx = 0 #Define index as 0
for line in lines: # For loop through the lines
    if text_search in line: #If the text_search (inputted keyword) appears in lines
        new_list.insert(idx, line) #Insert the line with its corresponding index into the new list 
        idx = idx + 1 #Adds 1 to the index of each presence of the keyword
file_read.close() #Close the file
```

# Print out the search result

```python
if len(new_list) == 0: #If the length of new list is 0, meaning no text_search is found
    print("'" + text_search + "'" + " is not presented in the file.") #Print this statement
elif len(new_list) > 0: #Else if the length of new list is bigger than 0, meaning text_search is found
    print("'" + text_search + "'" + " is presented in following lines of the file for " + str(idx) + " times.\n\n") #Print this statement including the index as indicating the times of presence
    for x in range(len(new_list)): 
        print(new_list[x]) #Print the list in each lines
```
