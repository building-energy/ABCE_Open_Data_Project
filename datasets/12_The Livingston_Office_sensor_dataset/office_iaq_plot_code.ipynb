# Example of data analysis

This notebook shows how to analyse the office iaq sensor data example from the ABCE Open Data Project.

View this dataset and notebook on GitHub here: https://github.com/building-energy/ABCE_Open_Data_Project/tree/main/internal_test_datasets/simulation

Import all the related libraries

```python
import pandas as pd
import requests
from matplotlib import pyplot as plt  
```

To obtain csv file from the github url add "?raw=true" after .csv and then copy the link to the python code

```python
csv_download_url = 'https://raw.githubusercontent.com/building-energy/ABCE_Open_Data_Project/main/external_datasets/office_iaq_sensor_data/office_livingstone_aq_1.csv'
df = pd.read_csv(csv_download_url, index_col = 0, parse_dates = True)
```

Now we run the code to create the comfort and IAQ plot of the office space

```python
plt.rcParams["figure.figsize"] = [15.00, 7.0]
plt.rcParams["figure.autolayout"] = True
fig, ax1 = plt.subplots()
ax1.plot(df["temperature ['oC']"], color='red', label ='Indoor air temperature ($^\circ$C)')
ax1.set_ylabel('Indoor air temperature ($^\circ$C)', color='red')
ax2 = ax1.twinx()
ax2.plot(df["co2 ['ppm']"], color='blue', label ='CO2 concentration (ppm)')
ax2.set_ylabel('CO2 concentration (ppm)', color='blue' )
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right',)
plt.show()
```
